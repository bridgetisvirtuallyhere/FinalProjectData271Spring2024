### Dataset Significance:

- Understanding housing element compliance is crucial for meeting the housing needs of Californians and ensuring that local governments fulfill their planning obligations.
- Compliance with housing element law is essential for accessing state funding programs for transportation, infrastructure, and housing.
- The dataset provides transparency and accountability by documenting the compliance status of housing elements submitted by jurisdictions.
- It assists policymakers, urban planners, and stakeholders in assessing the effectiveness of housing planning efforts and addressing housing challenges in California.

IN compliance with State housing element law means that the housing plan submitted by the jurisdiction meets the requirements set forth by the State of California.
1. **Meeting Housing Needs**: The housing plan adequately addresses the current and projected housing needs of the community, including housing for various income levels and demographic groups.

2. **Alignment with State Regulations**: The housing plan aligns with state regulations and guidelines governing housing element submissions. This includes adherence to statutory requirements outlined by the California Department of Housing and Community Development (HCD).

3. **Legal and Regulatory Compliance**: The housing plan complies with all relevant laws, regulations, and policies related to housing planning, zoning, and land use.

4. **Approval for Implementation**: Approval of a compliant housing element by HCD allows the jurisdiction to implement the housing plan as part of its general plan. This enables the jurisdiction to pursue funding opportunities, implement housing policies, and guide future development decisions.

5. **Access to Funding**: A compliant housing element enhances the jurisdiction's eligibility for state funding programs aimed at supporting housing, transportation, infrastructure, and community development initiatives.

6. **Long-term Planning**: A compliant housing element reflects the jurisdiction's commitment to long-term planning and addressing housing challenges in a proactive and sustainable manner.


OUT means:
1. **Non-compliance with State Law**: The housing plan submitted by the jurisdiction does not meet the requirements set forth by State housing element law. This could be due to various factors such as insufficient housing projections, inadequate zoning for housing development, or failure to address the housing needs of diverse income groups.

2. **Missing Submission**: It may indicate that the jurisdiction has not yet submitted the adopted housing element for review by the California Department of Housing and Community Development (HCD) within the required timeframe.

3. **Review Pending**: In some cases, "Out" status may also mean that the housing element is currently under review by HCD, and a determination of compliance has not been finalized.

4. **Compliance Issues**: The housing plan may have been reviewed by HCD, and compliance issues have been identified, necessitating revisions or updates to meet statutory requirements.

5. **Implications for Funding**: Non-compliance with housing element law may impact the jurisdiction's eligibility for certain state funding programs that require a compliant housing element as a prerequisite for funding allocation.

### Insights:

1. **County**:
   - Los Angeles County has the highest number of housing elements submitted, followed by Orange County and Riverside County.
   - Some counties have very few housing elements submitted, indicated by a count of 1 or 2.

2. **Jurisdiction**:
   - There are a total of 539 unique jurisdictions submitting housing elements.
   - Each jurisdiction typically submits only one housing element, as indicated by the count of 1 for most jurisdictions.

3. **Planning Period**:
   - The majority of housing elements cover a planning period of 6 years.
   - Only a small proportion cover a planning period of 7 years.

4. **Record Type**:
   - The most common record type is 'Adopted', indicating housing elements that have been officially adopted by the jurisdictions.
   - There are also housing elements categorized as 'Subsequent Draft', 'Initial Draft', 'New Cycle', and 'Draft', though these are less common.

5. **Review Status**:
   - The majority of housing elements are either 'In' or 'Out' of review, indicating whether they have been reviewed by the California Department of Housing and Community Development (HCD).
   - Some housing elements are 'In Review', indicating they are currently undergoing review by HCD, while others are 'Due', meaning they are awaiting review.

6. **Compliance Status**:
   - Most housing elements have a compliance status of 'IN', indicating compliance with State housing element law.
   - A smaller proportion have a compliance status of 'OUT', indicating non-compliance or pending submission.

```python
# calculate the time taken for review
ca_co_df['Review Duration'] = ca_co_df['Date Reviewed'] - ca_co_df['Date Received']
```

```python
ca_co_df.head()
```

```python
plt.figure(figsize=(5, 12))
sns.countplot(data=ca_co_df, y='Review Duration', order=ca_co_df['Review Duration'].value_counts().index)
plt.title('Boxplot of Review Duration')
plt.ylabel('Review Duration')
plt.show()
```

Seems like it either mostly takes above 50 days or 0 days.

```python
# Convert 'Review Duration' to integer type
ca_co_df['Review Duration'] = ca_co_df['Review Duration'].dt.days.astype(int)

# Visualize using a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=ca_co_df, y='Review Duration')
plt.title('Violin Plot of Review Duration')
plt.ylabel('Review Duration (days)')
plt.show()
```

Looks like 3 distributions, but what I said earlier holds up, either >50 days or 0 day in review.
